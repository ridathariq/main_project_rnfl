```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# ===============================
# Mean-Zero + Dynamic Contrast Enhancement
# ===============================

def mean_zero_dce(img):
    """
    Perform Mean-Zero Normalization + Dynamic Contrast Enhancement (Gamma + Unsharp Mask)
    for retinal preprocessing (RNFL / PPA analysis).
    """
    # 1. Extract green channel (best contrast for retina)
    green = img[:, :, 1].astype(np.float32)

    # 2. Mean-zero normalization
    mean_val = np.mean(green)
    img_zero = green - mean_val

    # 3. Scale to 0–255 range
    min_val = np.min(img_zero)
    max_val = np.max(img_zero)
    img_norm = (img_zero - min_val) / (max_val - min_val)
    img_norm = (img_norm * 255).astype(np.uint8)

    # 4. DCE Step 1 — Gamma correction
    gamma = 1.2   # adjustable: 1.1–1.5 usually works well
    img_gamma = np.power(img_norm / 255.0, gamma)
    img_gamma = (img_gamma * 255).astype(np.uint8)

    # 5. DCE Step 2 — Unsharp Mask for edge enhancement
    blur = cv2.GaussianBlur(img_gamma, (0, 0), sigmaX=3)
    unsharp = cv2.addWeighted(img_gamma, 1.5, blur, -0.5, 0)

    return unsharp

# ===============================
# Load Image
# ===============================

# Replace your image name here
img = cv2.imread("retina.jpg")

# If image not found
if img is None:
    raise FileNotFoundError("Image not found! Place retina.jpg in the same folder.")

# Convert BGR → RGB (for plotting only)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# ===============================
# Apply preprocessing
# ===============================

processed = mean_zero_dce(img)

# ===============================
# Display Results
# ===============================

plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img_rgb)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Mean-Zero + DCE Output")
plt.imshow(processed, cmap='gray')
plt.axis("off")

plt.show()

# ===============================
# Save output
# ===============================

cv2.imwrite("mean_zero_dce_output.png", processed)

print("Saved output as mean_zero_dce_output.png")
```
